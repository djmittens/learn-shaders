```python
# %matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0., 5., num=500)
y = lambda x, k: np.abs(np.cos(k * np.pi * x))

    

plt.subplots_adjust(bottom=.25)
fig, ax = plt.subplots()
ax.grid(True)
ax.set_aspect('equal')

ax.axhline(y=0, color = 'k')
ax.axvline(x=0, color = 'k')


ax_slide = plt.axes([0.25, .1, .65, .03]) # xpos, ypos, width, height
s_factor = plt.Slider(ax_slide, 'Scale', .1, 6, valinit=6, valstep=.2)
p = ax.plot(x, y(x, 2))

def update(val):
    p.set_ydata(y(x,val))
    fig.canvas.draw()
s_factor.on_changed(update)
# plt.show()
```
